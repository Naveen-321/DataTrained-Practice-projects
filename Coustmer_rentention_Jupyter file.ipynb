```python
 #generate related variables
from numpy import mean
from numpy import std
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot
# seed random number generator
seed(1)
# prepare data
sheet_1 = 20 * randn(1000) + 100
sheet_2 = sheet_1 + (10 * randn(1000) + 50)
# summarize
print('sheet_1: mean=%.3f stdv=%.3f' % (mean(sheet_1), std(sheet_1)))
print('sheet_2: mean=%.3f stdv=%.3f' % (mean(sheet_2), std(sheet_2)))
# plot
pyplot.scatter(sheet_1, sheet_2)
pyplot.show()
```

```python
# calculate the covariance between two variables
from numpy.random import randn
from numpy.random import seed
from numpy import cov
from matplotlib import pyplot

# seed random number generator
seed(1)

# prepare data
sheet_1 = 20 * randn(1000) + 100
sheet_2 = sheet_1 + (10 * randn(1000) + 50)

# calculate covariance matrix
covariance = cov(sheet_1, sheet_2)
print(covariance)

from scipy.stats import pearsonr
# prepare data
sheet_1 = 20 * randn(1000) + 100
sheet_2 = sheet_1 + (10 * randn(1000) + 50)


# calculate Pearson's correlation
corr, _ = pearsonr(sheet_1, sheet_2)
print('Pearsons correlation: %.3f' % corr)

```

```python
from scipy.stats import spearmanr
# seed random number generator
seed(1)
# prepare data
sheet_1 = 20 * randn(1000) + 100
sheet_2 = sheet_1 + (10 * randn(1000) + 50)


# calculate spearman's correlation
corr, _ = spearmanr(sheet_1, sheet_2)
print('Spearmans correlation: %.3f' % corr)
```
